```python
#importing pandas to read csv file
import pandas as pd

school = pd.read_csv ('schools.csv')

#setting threshold to determine the best math result
threshold = 800 * 0.8

#Which NYC schools have the best math results?
math_schools = school[school['average_math'] >= threshold][['school_name', 'average_math']]
best_math_schools = math_schools.sort_values(by='average_math', ascending = False)
print(best_math_schools)

#total scores per school
total_SAT = (school['average_math'] + school['average_reading'] + school['average_writing']).rename('total_SAT')
school_scores = pd.concat([math_schools['school_name'], total_SAT], axis = 1)

#top 10 performing schools based on the combined SAT scores
top_10_schools = school_scores.sort_values(by = 'total_SAT', ascending = False).head(10) 
print(top_10_schools)
#adding total_SAT as a column in school
school['total_SAT'] = total_SAT
borough_stats = school.groupby("borough")["total_SAT"].agg(
    num_schools="count",
    average_SAT="mean",
    std_SAT="std"
)

# round values
borough_stats = borough_stats.round(2)

#largest standard deviation in the combined SAT score
largest_std_dev = borough_stats.sort_values("std_SAT", ascending=False).head(1)
print(largest_std_dev)
```
